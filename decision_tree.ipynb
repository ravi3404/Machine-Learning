```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df[' education']= le.fit_transform(df[' education'])
```

```python
df.head()
```

```python
# df[' self_employed']= pd.get_dummies(df[' self_employed'])
```

```python
# df = pd.get_dummies(df, columns=[' self_employed'])   
cat_cols = [' education', ' self_employed']
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
dtc = DecisionTreeClassifier( max_depth = 4, random_state = 50)
X = df.drop(columns = [' loan_status'], axis=1)
Y = df[' loan_status']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 50)

dtc.fit(X_train,Y_train)
```

```python
pred_y = dtc.predict(X_test)
```

```python
print("Accuracy score :", accuracy_score(Y_test,pred_y))
```

```python
classification_rep = classification_report(Y_test, pred_y)
print(classification_rep)
```

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dtc, filled=True, feature_names=X.columns, class_names=['0', '1'])
plt.show()
```
